```python
φ_1 = 180 - 55 # degrees.  Python trig functions are in radians so need to convert from degrees to radians.
print(f"φ = {φ_1}° = {radians(φ_1):0.2f} rads")
φ_1 = radians(φ_1) # convert φ_1 to radians
F_1z = F_1 * cos(φ_1) # lbf
print(f"F_1z = {F_1z:0.2f} lbf")
```

Find the "horizontal" component $F_{1h}$, that is, in the `xy` plane.  Think of this as the projection of the $F_{1}$ force onto the `xy` plane.

```python
F_1h = F_1 * sin(φ_1) # lbf
print(f"F_1h = {F_1h:0.2f} lbf")
```

Split this into the `x` and `y` components.  Find the `x` component first.  By looking at the angles of $F_{1}$ it seems that the `x` and `z` components should be negative, `y` positive. Also expect the magnitude of each component to be greatest in y, z, then x.  If you are asking why I chose `cos()` versus `sin()`...  That's why I draw the graph of `sin()` and `cos()` functions above.  If θ<sub>1</sub> is 0 then $F_{1x}$ would be the greatest positive value and if it were 90° then $F_{1x}$ would be 0.  This corresponds to `cos()`.  Rather than try to memorize formulas try to think intuitively about what's happening.

```python
θ_1 = 90 + 30 # degrees.  This is the angle from the x-axis. The x-component should be negative and the smallest.
print(f"θ_1 = {θ_1}° = {radians(θ_1):0.2f} rads")
θ_1 = radians(θ_1) # radians
F_1x = F_1h * cos(θ_1) # lbf
print(f"F_1x = {F_1x:0.2f} lbf")
```

Similarly, find $F_{1y}$.

```python
F_1y = F_1h * sin(θ_1) # lbf
print(f"F_1y = {F_1y:0.2f} lbf")
```

See if our values turned out like we supposed they would.  That is, x and z negative, y positive, and magnitude in order of: y, z, x.

```python
print(f"{F_1x:0.2f} < 0 is {F_1x < 0}, {F_1y:0.2f} > 0 is {F_1y > 0}, {F_1z:0.2f} < 0 is {F_1z < 0}")
print(f"|{F_1y:0.2f}| > |{F_1z:0.2f}| > |{F_1x:0.2f}| == {abs(F_1y) > abs(F_1z) > abs(F_1x)}")
```

Woot!  But, **note** we disagreed (swapped) on the values of $F_{1x}$ and $F_{1y}$.  I'm going to stand behind my thinking that $F_{1x}$ is going to be smaller than $F_{1y}$ because $F_{1y}$ is getting more of that $F_{1xy}$ force than $F_{1x}$.

**Find the components of $F_{2}$.**  

Just looking at it (hard to "see" a 3D projection on a 2D plane), I would guess greatest magnitude are in order of `z`, `x`, and `y`, but can't really see if `x` or `y` will be greater.  `z` and `x` will be positive but it's actually impossible to tell if `y` is positive or negative. Looking at the drawing I guess it's in the negative `y` direction but someone else could just as easily say it's in the positive `y` direction. Since I'm assuming it's in the negative `y` direction will have a negative `y` value.

The $F_{2x}$ and $F_{2z}$ components are pretty trivial to get:

```python
θ_2 = 65 # degrees
print(f"θ_2 = {θ_2}° = {radians(θ_2):0.2f} rads")
θ_2 = radians(θ_2) # Convert to radians.

F_2x = F_2 * cos(θ_2) # Again, we know it's cos() because we looked at the cos() function above.
print(f"F_2x = {F_2x:0.2f} lbf")
```

```python
γ_2 = 32 # degrees
print(f"θ_2 = {γ_2}° = {radians(γ_2):0.2f} rads")
γ_2 = radians(γ_2) # Convert to radians.

F_2z = F_2 * cos(γ_2) # Again, we know it's cos() because we looked at the cos() function above.
print(f"F_2z = {F_2z:0.2f} lbf")
```

To get $F_{2y}$ we can use the formula $F_{2}^2 = F_{2x}^2 + F_{2y}^2 + F_{2z}^2$ since we already have three of the four values.  This also shows that we still can't determine if $F_{2y}$ is positive or negative since both values will satisfy this equation.  Your prof underspecified the input. Python's not made for symbolic math so we manually rerrange the above equation to solve for $F_{2y}$:  
$F_{2y} = \pm\sqrt{F_{2}^2 - F_{2x}^2 - F_{2z}^2}$

```python
F_2y = sqrt(F_2**2 - F_2x**2 - F_2z**2) # lbf
print(f"F_2y = ±{F_2y:0.2f} lbf")
```

To me it looks like $F_{2y}$ is going in the negative direction so will force it to be negative here.

```python
F_2y = -sqrt(F_2**2 - F_2x**2 - F_2z**2) # lbf
print(f"F_2y = {F_2y:0.2f} lbf")
```

See if our earlier assumptions held up. $\lvert $F_{2z}$ \rvert > \lvert $F_{2x}$ \rvert > \lvert $F_{2y}$ \rvert$, and $F_{2x}$ and $F_{2x}$ positive.

```python
print(f"|{F_2z}| > |{F_2x}| > |{F_2y}| == {abs(F_2z) > abs(F_2x) > abs(F_2y)}")
```

**Find the components of $F_{2}$.**  
Since we are given a point on the line it will be trivial to convert to a unit vector and call it a day.  
Pretty obvious the relative magnitude and signs of the resultant force vector $F_{3}$.

The unit vector is $\hat{u} =({\frac {u_{x}}{\|\mathbf {u} \|}},{\frac {u_{y}}{\|\mathbf {u} \|}},{\frac {u_{z}}{\|\mathbf {u} \|}})$

```python
x_3, y_3, z_3 = 6, 2, -2 # distance units
abs_u_3 = sqrt(x_3**2 + y_3**2 + z_3**2)
u_3 = (x_3/abs_u_3, y_3/abs_u_3, z_3/abs_u_3) # Unit vectors are dimensionless.
print(f"unit vector u_3 = [{', '.join(f'{u:0.2f}{d}' for u,d in zip(u_3, dirs))}]") 
```

Double check this is in fact a unit vector:

```python
print(sum(u**2 for u in u_3)**0.5 == 1)
```

To calculate the force vector it's just the magnitude of the force times the unit vector:

```python
F_3x, F_3y, F_3z = [F_3 * u for u in u_3] # lbf
print(f"F_3x, F_3y, F_3z = ({F_3x:0.2f}i, {F_3y:0.2f}j, {F_3z:0.2f}k) lbf")
```

Trivial to see that the signs are correct and the relative magnitudes are `[3, 1, 1]` as expected.

Collate all the above work into vectors:

```python
F_1v = (F_1x, F_1y, F_1z)
F_2v = (F_2x, F_2y, F_2z)
F_3v = (F_3x, F_3y, F_3z)
```

```python
for i, Fv in enumerate([F_1v, F_2v, F_3v]):
    print(f"F_{i+1}v = ({', '.join(f'{F:0.2f}{d}' for F,d in zip(Fv, dirs))}) lbf")
```

The resultant force vector is the sum of these:

```python
F_v = [sum(F) for F in list(zip(F_1v, F_2v, F_3v))] # lbf
print(f"F_v = ({', '.join(f'{F:0.2f}{d}' for F,d in zip(F_v, dirs))}) lbf")
```

Convert this to the force magnitude times a unit vector (which I guess is the same as "directional cosines").  Use the exact same method as we did before.

```python
F_mag = sqrt(sum(F**2 for F in F_v)) # lbf
F_uv = [F/F_mag for F in F_v]
F_uv_str = ", ".join([f"{(F):0.2f}{d}" for F,d in zip(F_uv, dirs)]) # Convert to formatted string.
print(f"The resultant magnitude is {F_mag:0.2f} lbf and the unit vector is ({F_uv_str}).")
```

Convert the unit vector to angles:

```python
for uv, label in zip(F_uv, ["α", "β", "γ"]):
    print(f"{label} = acos({uv:0.2f}) = {acos(uv):0.2f} radians = {degrees(acos(uv)):0.2f}°")
```

These represent the angles from each of the axis x, y, and z.
